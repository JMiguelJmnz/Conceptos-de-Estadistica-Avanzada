La variable que mas se relaciona con el precio de venta es la calidad general (OverallQual) con 79%

```python
import statsmodels.api as sm

X = numeric_df.drop("SalePrice", axis=1)
y = numeric_df["SalePrice"]
```

```python
X
```

```python
y
```

```python
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
valores_p = model.pvalues
variables_significativas = valores_p[valores_p < 0.05].index
var_sig = X[variables_significativas]
var_sig.head()
```

```python
var_sig.describe()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
var_sig_con_constante = sm.add_constant(var_sig)

vif_data = pd.DataFrame()
vif_data["variable"] = var_sig_con_constante.columns
vif_data["VIF"] = [variance_inflation_factor(var_sig_con_constante.values, i) for i in range(var_sig_con_constante.shape[1])]

print(vif_data)
```

Las variables que presentan multi colinealidad, son 1stFlrSF, 2ndFlrSF, GrLivArea

```python
X_filtrada = var_sig_con_constante.drop(["1stFlrSF", "2ndFlrSF", "GrLivArea"], axis=1)

model = sm.OLS(y, X_filtrada).fit()
print(model.summary())
```

Con todo el anÃ¡lisis anterior construye el modelo de regresiÃ³n final excluyendo aquellas variables que consideres necesarias y mide el nivel de precisiÃ³n del modelo por medio del error cuadratico medio y la R cuadrada. Â¿QuÃ© tan preciso fue tu modelo?

```python
rss = sum((model.fittedvalues - y) ** 2)  
df_residual = len(y) - X.shape[1] 
mse = rss / df_residual
print(f'Error cuadratico medio (MSE): {mse}')
```
